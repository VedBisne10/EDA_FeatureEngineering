```python
# Feature engineering on date of journey
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

df['Date']
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
# Drop date of journey
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Feature engineering on arrival time
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0]) #It separated the hour-minutes and date-month from each other giving only hour-min because index value is 0
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time', axis = 1, inplace=True)
```

```python
df.head(2)
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
df.head(2)
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)
df.info()
```

```python
df.head(2)
```

```python
df['Duration']
```

```python
df['Duration_hour'] = df['Duration'].str.split(' ').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1]
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Duration_hour'] = (df['Duration'].str.extract(r'(\d+)h')[0].fillna(0).astype(int))
df['Duration_min'] = (df['Duration'].str.extract(r'(\d+)m')[0].fillna(0).astype(int))
```
